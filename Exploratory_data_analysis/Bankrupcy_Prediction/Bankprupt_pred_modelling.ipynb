```python
col_dict = {}

for col in df.columns:
    col_dict[col] = col.strip()

df.rename(col_dict, axis='columns', inplace=True)
```

Checking for zero variance columns

```python
zero_variance_column = []

for col in df.columns:
    if(df[col].nunique() == 1):
        zero_variance_column.append(col)
 
print(zero_variance_column)
```

Checking null values

```python
missing_data = df.isnull().mean()*100
missing_data = missing_data[missing_data>0]

print(missing_data)
```

Some important Visualisations

```python
# Create a line plot with seaborn
sns.lineplot(x='ROA(C) before interest and depreciation before interest', y='ROA(A) before interest and % after tax', data=df, color='red')
sns.lineplot(x='ROA(C) before interest and depreciation before interest', y='ROA(B) before interest and depreciation after tax', data=df, color='green')

# Set the title and labels
plt.title('Return on Assets')
plt.xlabel('ROA(C) before interest and depreciation before interest')
plt.ylabel('ROA(A), ROA(B)')

# Show the plot
plt.show()
```

```python
# Extract the relevant columns
net_value_b = df['Net profit before tax/Paid-in capital'][df['Bankrupt'] == 1]
net_value_a = df['Net profit before tax/Paid-in capital'][df['Bankrupt'] == 0]

plt.figure(figsize=(10, 6))

sns.lineplot(net_value_a, label='Healty', color = 'gray')
sns.lineplot(net_value_b, label='Bankrupt', color = 'red')

# Add a legend
plt.ylabel('Net profit before tax/Paid-in Capital')
plt.legend()

# Show the plot
plt.show()
```
