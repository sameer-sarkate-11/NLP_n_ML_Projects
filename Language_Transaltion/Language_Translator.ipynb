```python
import os
import sys
import transformers
import tensorflow as tf
from datasets import load_dataset
from transformers import AutoTokenizer
from transformers import TFAutoModelForSeq2SeqLM, DataCollatorForSeq2Seq
from transformers import AdamWeightDecay
```

```python
ds = load_dataset("cfilt/iitb-english-hindi")
```

```python
ds
```

```python
ds['train']['translation']
```

```python
%pip install datasets
import random
from datasets import Dataset
```

```python
random_rows = random.sample(ds['train']['translation'], int(len(ds['train']['translation'])/150))
```

```python
new_d = {
    'translation' : random_rows,
}
```

```python
train = Dataset.from_dict(new_d)
train
```

```python
model_name = "Helsinki-NLP/opus-mt-en-hi"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = TFAutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
tokenizer("This is a demo sentence")
```

```python
with tokenizer.as_target_tokenizer():
  print(tokenizer("एक्सेर्साइसर पहुंचनीयता अन्वेषक"))
```

```python
max_len = 128

def preproc(samples):
  inputs = [sample["en"] for sample in samples["translation"]]
  targets = [sample["hi"] for sample in samples["translation"]]

  model_inputs = tokenizer(inputs, max_length = max_len, truncation = True)

  with tokenizer.as_target_tokenizer():
    labels = tokenizer(targets, max_length = max_len, truncation = True)

  model_inputs["labels"] = labels["input_ids"]
  return model_inputs
```

```python
tokenised_data = ds.map(preproc, batched=True)
```

```python
tokenised_train = train.map(preproc, batched=True)
```

```python
batch_size = 16
learning_rate = 2e-5
weight_decay = 0.01
```

```python
data_collator = DataCollatorForSeq2Seq(tokenizer, model=model, return_tensors = "tf")
```

```python

train_dataset = model.prepare_tf_dataset(
    tokenised_train,
    batch_size = batch_size,
    shuffle = True,
    collate_fn = data_collator
)
```

```python
validation_dataset = model.prepare_tf_dataset(
    tokenised_data["validation"],
    batch_size = batch_size,
    shuffle = False,
    collate_fn = data_collator
)
```

```python
generation_dataset = model.prepare_tf_dataset(
    tokenised_data["validation"],
    batch_size = batch_size,
    shuffle = True,
    collate_fn = data_collator
)
```

```python
tf.config.run_functions_eagerly(True)
```

```python
optimizer = AdamWeightDecay(learning_rate = learning_rate, weight_decay_rate = weight_decay)
model.compile(optimizer=optimizer, run_eagerly=True)
```

```python
model.fit(train_dataset, validation_data = validation_dataset, epochs = 2)
```

```python
model.save_pretrained("tf_model/")
```

```python
input_text = "India is a great country"

tokenized_input = tokenizer([input_text], return_tensors='np')
out = model.generate(**tokenized_input, max_length = 128)

print(out)
```
